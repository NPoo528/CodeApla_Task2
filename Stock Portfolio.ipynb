```python
import yfinance as yf
import pandas as pd

# Initialize an empty portfolio
portfolio = {}

# Function to add stock to the portfolio
def add_stock(ticker, shares):
    if ticker in portfolio:
        portfolio[ticker] += shares
    else:
        portfolio[ticker] = shares
    print(f"Added {shares} shares of {ticker} to the portfolio.")

# Function to remove stock from the portfolio
def remove_stock(ticker, shares):
    if ticker in portfolio:
        if portfolio[ticker] > shares:
            portfolio[ticker] -= shares
            print(f"Removed {shares} shares of {ticker} from the portfolio.")
        elif portfolio[ticker] == shares:
            del portfolio[ticker]
            print(f"Removed all shares of {ticker} from the portfolio.")
        else:
            print(f"Cannot remove {shares} shares of {ticker}. You only have {portfolio[ticker]} shares.")
    else:
        print(f"{ticker} is not in the portfolio.")

        # Function to get current stock prices
def get_stock_prices(tickers):
    prices = {}
    for ticker in tickers:
        stock = yf.Ticker(ticker)
        prices[ticker] = stock.history(period='1d')['Close'].iloc[-1]
    return prices

# Function to calculate portfolio value
def calculate_portfolio_value(portfolio, prices):
    total_value = 0
    portfolio_details = []

    for ticker, shares in portfolio.items():
        stock_value = shares * prices[ticker]
        total_value += stock_value
        portfolio_details.append({
            'Ticker': ticker,
            'Shares': shares,
            'Price': prices[ticker],
            'Total Value': stock_value
        })

    return total_value, pd.DataFrame(portfolio_details)

# Function to display portfolio
def display_portfolio():
    if not portfolio:
        print("The portfolio is empty.")
        return

    tickers = list(portfolio.keys())
    stock_prices = get_stock_prices(tickers)
    total_value, portfolio_df = calculate_portfolio_value(portfolio, stock_prices)

    print("Current Portfolio Value: ${:.2f}".format(total_value))
    display(portfolio_df)

# Example usage
add_stock('AAPL', 10)  # Add 10 shares of Apple
add_stock('GOOGL', 5)  # Add 5 shares of Alphabet
add_stock('AMZN', 2)
remove_stock('AAPL', 5)  # Remove 5 shares of Apple
display_portfolio()  # Display the portfolio
```
